# <b><span style="color:red">Image Augmentation</span></b>

#### This notebook explores different experiment with augmentation techniques in Torch

```python
from pathlib import Path
import torch
import matplotlib.pyplot as plt
from torchvision.transforms import v2
from torchvision.io import read_image
import os
torch.manual_seed(1)

imgPath=(os.path.join('../project_files','data','AlexBrush','image_0.png'))
img = read_image(imgPath)
```

```python
def plot_image_with_output(img, out1,out2,out3):
    plt.figure(figsize=(10, 5))
    plt.subplot(2, 2, 1)
    plt.imshow(img.permute(1, 2, 0))
    plt.subplot(2, 2, 2)
    plt.imshow(out1.permute(1, 2, 0))
    plt.subplot(2, 2, 3)
    plt.imshow(out2.permute(1, 2, 0))
    plt.subplot(2, 2, 4)
    plt.imshow(out3.permute(1, 2, 0))
    plt.show()
```

```python
#random crop augmentaion
transform = v2.RandomCrop(size=(30, 204))
out1 = transform(img)
out2 = transform(img)
out3= transform(img)
#display the image and the transformed image
plot_image_with_output(img, out1,out2,out3)
```

```python
#combinig the random crop and horizontal flip
transforms = v2.Compose([
    v2.RandomResizedCrop(size=(30, 224), antialias=True),
    v2.RandomHorizontalFlip(p=0.5),
 
])
out1 = transforms(img)
out2 = transforms(img)
out3 = transforms(img)
#display the image and the transformed image
plot_image_with_output(img, out1,out2,out3)
```

```python
#apply scalejitter to the image
transforms = v2.Compose([
    v2.ScaleJitter((50,50))
])
out1 = transforms(img)
out2 = transforms(img)
out3= transforms(img)
plot_image_with_output(img, out1,out2,out3)
```

```python
#apply randomshortestsize to the image
transform =v2.RandomShortestSize(50,50)
out1 = transform(img)
out2 = transform(img)
out3= transform(img)
plot_image_with_output(img, out1,out2,out3)
```

```python
#apply centercrop to the image
transforms = v2.Compose([
    v2.CenterCrop(size=(40, 100))
])
out1 = transforms(img)
out2 = transforms(img)
out3= transforms(img)
plot_image_with_output(img, out1,out2,out3)
```

```python
#apply randomzoomout to the image
transforms = v2.Compose([
    v2.RandomZoomOut(fill=255)
])
out1 = transforms(img)
out2 = transforms(img)
out3= transforms(img)
plot_image_with_output(img, out1,out2,out3)
```

```python
#apply randomrotation to the image
transforms = v2.Compose([
    v2.RandomRotation(15, expand=True),
    v2.ColorJitter(brightness=(1,15)),
     v2.GaussianBlur(5),
     v2.RandomAdjustSharpness(5),
     v2.ToDtype(torch.float32, scale=True),
     v2.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
out1 = transforms(img)
out2 = transforms(img)
out3= transforms(img)
plot_image_with_output(img, out1,out2,out3)
```

```python
#apply randomaffine to the image
transforms = v2.Compose([
    v2.RandomAffine(20)
])
out1 = transforms(img)
out2 = transforms(img)
out3= transforms(img)
plot_image_with_output(img, out1,out2,out3)
```

```python
#apply randomperspective to the image
transforms = v2.Compose([
    v2.RandomPerspective(0.5)
])
out1 = transforms(img)
out2 = transforms(img)
out3= transforms(img)
plot_image_with_output(img, out1,out2,out3)
```

```python
#apply elastic transform to the image
transforms = v2.Compose([
    v2.ElasticTransform(15)
])
out1 = transforms(img)
out2 = transforms(img)
out3= transforms(img)
plot_image_with_output(img, out1,out2,out3)
```

```python
#apply colorjitter to the image
transforms = v2.Compose([
    v2.ColorJitter(brightness=(1,10),contrast=(1,5))
])
out1 = transforms(img)
out2 = transforms(img)
out3= transforms(img)
plot_image_with_output(img, out1,out2,out3)
```

```python
#randomphotometricdistort to the image
transforms = v2.Compose([
    v2.RandomPhotometricDistort(brightness=(0.5,15))
])
out1 = transforms(img)
out2 = transforms(img)
out3= transforms(img)
plot_image_with_output(img, out1,out2,out3)
```

```python
#apply gaussianblur to the image
transforms = v2.Compose([
    v2.GaussianBlur(11)
])
out1 = transforms(img)
out2 = transforms(img)
out3= transforms(img)
plot_image_with_output(img, out1,out2,out3)
```

```python
transforms = v2.Compose([
    v2.RandomPosterize(8)
])
out1 = transforms(img)
out2 = transforms(img)
out3= transforms(img)
plot_image_with_output(img, out1,out2,out3)
```

```python
#0 for blur,1 for original image, >1 for sharpen
transforms = v2.Compose([
    v2.RandomAdjustSharpness(10)
])
out1 = transforms(img)
out2 = transforms(img)
out3= transforms(img)
plot_image_with_output(img, out1,out2,out3)
```

```python
transforms = v2.Compose([
    v2.RandomAutocontrast(0.5)
])
out1 = transforms(img)
out2 = transforms(img)
out3= transforms(img)
plot_image_with_output(img, out1,out2,out3)
```
